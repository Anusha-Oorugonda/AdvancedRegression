### Problem Statement

A US-based housing company named Surprise Housing has decided to enter the Australian market. The company uses data analytics to purchase houses at a price below their actual values and flip them on at a higher price. 

**The company wants to know**:

- Which variables are significant in predicting the price of a house, and
- How well those variables describe the price of a house.


**Business Goal**
- Need to create model based on the price of houses with the available independent variables. 
- This model will then be used by the management to understand how exactly the prices vary with the variables.
- They can accordingly manipulate the strategy of the firm and concentrate on areas that will yield high returns. 
- Further, the model will be a good way for management to understand the pricing dynamics of a new market.
- determine the optimal value of lambda for ridge and lasso regression.
- find the top most influencing parameters which describes the price of the house and yield more returns

As we found above,likewise there are many columns which are explaining lowest variance,we will write a method to get the column names that have count of one value more than 85%

```python
# Method to get the column names that have count of one value more than 85%

def getHighCategoricalValueCounts():
    column = []
    categorical_columns = housing_price.select_dtypes(include=['object'])
    for col in (categorical_columns):
        if(housing_price[col].value_counts().max() >= 1241):
            column.append(col)
    return column

columnsToBeRemoved = getHighCategoricalValueCounts()

# Remove the columns with skewed data
print(columnsToBeRemoved)
housing_price.drop(columnsToBeRemoved, axis = 1, inplace = True)

housing_price.head()
```

```python
num_col = ['LotArea','MasVnrArea','BsmtFinSF1','BsmtFinSF2','TotalBsmtSF','1stFlrSF','GrLivArea','OpenPorchSF',
           'EnclosedPorch','3SsnPorch',
           'ScreenPorch' ,'PoolArea','MiscVal','SalePrice']
def drop_outliers(x):
    list = []
    for col in num_col:
        Q1 = x[col].quantile(.25)
        Q3 = x[col].quantile(.99)
        IQR = Q3-Q1
        x =  x[(x[col] >= (Q1-(1.5*IQR))) & (x[col] <= (Q3+(1.5*IQR)))] 
    return x   

housing_price = drop_outliers(housing_price)
```

### Step 5: Train Test Split

```python
#Creating train and test dataset for validation purpose
df_train,df_test = train_test_split(housing_price,train_size=0.7,test_size=0.3,random_state=42)
```

```python
housing_price[['ExterQual','BsmtQual','BsmtExposure','BsmtFinType1', 'HeatingQC','KitchenQual','GarageFinish','LotShape']].head()
```

**Scaling the train and test dataset and applying standard scaler**

```python
#Scaling the train and test dataset
num_col = ['MSSubClass','LotArea','OverallQual','OverallCond',
           'MasVnrArea','BsmtFinSF1',
           'BsmtFinSF2','BsmtUnfSF','TotalBsmtSF','1stFlrSF','2ndFlrSF',
           'LowQualFinSF','GrLivArea','BsmtFullBath','BsmtHalfBath','FullBath','HalfBath','BedroomAbvGr',
           'KitchenAbvGr','TotRmsAbvGrd','Fireplaces','GarageCars',
           'GarageArea','WoodDeckSF','OpenPorchSF','EnclosedPorch','3SsnPorch',
           'ScreenPorch','PoolArea','MiscVal','SalePrice']

scaler = StandardScaler()
df_train[num_col] = scaler.fit_transform(df_train[num_col])
df_test[num_col] = scaler.transform(df_test[num_col])
```

**Lets check the distribution after splitting and scaling**

```python
plt.figure(figsize=(16,6))
plt.subplot(121)
sns.distplot(df_train.SalePrice)
plt.subplot(122)
sns.distplot(df_test.SalePrice)
```

Spliting the dependent and independent variable

```python
y_train = df_train.pop('SalePrice')
X_train = df_train
```

```python
y_test = df_test.pop('SalePrice')
X_test = df_test
```

**Now using RFE lets try to to find the optimal number of feature,
Note : I cannot use RFE with GridSearchCV as 137 variable with 5 folds will create 685 fit and it will take more then 1 Hour to get the result so i am using RFE directly**

```python
print(len(X_train.columns))
print(X_train.shape)
```

**Now using RFE lets try to to find the optimal number of feature , lets get top 70 optimal independant features**

```python
lm  = LinearRegression()
lm.fit(X_train,y_train)
rfe = RFE(lm,70)
rfe.fit(X_train,y_train)
```

```python
rfe_scores = pd.DataFrame(list(zip(X_train.columns,rfe.support_,rfe.ranking_)))
rfe_scores.columns = ['Column_Names','Status','Rank']
```

```python
rfe_sel_columns = list(rfe_scores[rfe_scores.Status==True].Column_Names)
```

**Lets filter the train and test set for the RFE selected columns**

```python
X_train = X_train[rfe_sel_columns]
X_test = X_test[rfe_sel_columns]
X_train.head()
```

### Lasso regression model

```python
lm = Lasso(alpha=0.001)
lm.fit(X_train,y_train)

y_train_pred = lm.predict(X_train)
print(r2_score(y_true=y_train,y_pred=y_train_pred))

y_test_pred  = lm.predict(X_test)
print(r2_score(y_true=y_test,y_pred=y_test_pred))
```

```python
model_parameter = list(lm.coef_)
model_parameter.insert(0,lm.intercept_)
model_parameter = [round(x,3) for x in model_parameter]
col = df_train.columns
col.insert(0,'Constant')
list(zip(col,model_parameter))
```

**Now lets try to improve our model with the optimal value of alpha using GridSearchCV**

```python
folds = KFold(n_splits=10,shuffle=True,random_state=42)

hyper_param = {'alpha':[0.001, 0.01, 0.1,1.0, 5.0, 10.0,20.0]}

model = Lasso()

model_cv = GridSearchCV(estimator = model,
                        param_grid=hyper_param,
                        scoring='r2',
                        cv=folds,
                        verbose=1,
                        return_train_score=True
                       )

model_cv.fit(X_train,y_train)
```

```python
cv_result_l = pd.DataFrame(model_cv.cv_results_)
cv_result_l['param_alpha'] = cv_result_l['param_alpha'].astype('float32')
cv_result_l.head()
```

**Lets visualise alpha with mean test and train score of all the resulted models**

```python
plt.figure(figsize=(16,8))
plt.plot(cv_result_l['param_alpha'],cv_result_l['mean_train_score'])
plt.plot(cv_result_l['param_alpha'],cv_result_l['mean_test_score'])
plt.xscale('log')
plt.ylabel('R2 Score')
plt.xlabel('Alpha')
plt.show()
```

```python
# Checking the best parameter(Alpha value)
model_cv.best_params_
```

```python
lasso = Lasso(alpha=0.001)
lasso.fit(X_train,y_train)

y_train_pred = lasso.predict(X_train)
y_test_pred = lasso.predict(X_test)

print(r2_score(y_true=y_train,y_pred=y_train_pred))
print(r2_score(y_true=y_test,y_pred=y_test_pred))
```

```python
model_param = list(lasso.coef_)
model_param.insert(0,lasso.intercept_)
cols = df_train.columns
cols.insert(0,'const')
lasso_coef = pd.DataFrame(list(zip(cols,model_param)))
lasso_coef.columns = ['Feature','Coef']
```

```python
l_top10=lasso_coef.sort_values(by='Coef',ascending=False).head(10)
```

```python
#top10 predictors for house price are
l_top10=lasso_coef.sort_values(by='Coef',ascending=False).head(10)
plt.figure(figsize=(20,20))
plt.subplot(4,3,1)
sns.barplot(y = 'Feature', x='Coef', palette='Set1', data = l_top10)
plt.show()
```

```python
#top5 predictors are
l_top5=lasso_coef.sort_values(by='Coef',ascending=False).head(5)
plt.figure(figsize=(20,20))
plt.subplot(4,3,1)
sns.barplot(y = 'Feature', x='Coef', palette='Set1', data = l_top5)
plt.show()
```

### Ridge Regression

```python
ridge = Ridge(alpha=0.001)
ridge.fit(X_train,y_train)

y_train_pred = ridge.predict(X_train)
print(r2_score(y_train,y_train_pred))
y_test_pred = ridge.predict(X_test)
print(r2_score(y_test,y_test_pred))
```

**Now lets try to improve our model with the optimal value of alpha using GridSearchCV**

```python

folds  = KFold(n_splits=10,shuffle=True,random_state=42)

hyper_param = {'alpha':[0.001,0.01,0.1,0.2,0.5,0.9,1.0, 5.0, 10.0,20.0]}

model = Ridge()

model_cv = GridSearchCV(estimator=model,
                        param_grid=hyper_param,
                        scoring='r2',
                        cv=folds,
                        verbose=1,
                        return_train_score=True)

model_cv.fit(X_train,y_train)
```

```python
cv_result_r = pd.DataFrame(model_cv.cv_results_)
cv_result_r['param_alpha'] = cv_result_r['param_alpha'].astype('float32')
cv_result_r.head()
```

**Lets visualise alpha with R2 Score**

```python
plt.figure(figsize=(16,8))
plt.plot(cv_result_r['param_alpha'],cv_result_r['mean_train_score'])
plt.plot(cv_result_r['param_alpha'],cv_result_r['mean_test_score'])
plt.xlabel('Alpha')
plt.xscale('log')
plt.ylabel('R2 Score')
plt.show()
```

- From 0.9 the model is showing good improvement and R2 score is similar from 0.9 to 5 which is best so we will take alpha to be 5 from the below best params

```python
# Checking the best parameter(Alpha value)
model_cv.best_params_
```

```python
ridge = Ridge(alpha = 5)
ridge.fit(X_train,y_train)

y_pred_train = ridge.predict(X_train)
print(r2_score(y_train,y_pred_train))

y_pred_test = ridge.predict(X_test)
print(r2_score(y_test,y_pred_test))
```

```python
model_parameter = list(ridge.coef_)
model_parameter.insert(0,ridge.intercept_)
cols = df_train.columns
cols.insert(0,'constant')
ridge_coef = pd.DataFrame(list(zip(cols,model_parameter)))
ridge_coef.columns = ['Feature','Coef']
```

```python
ridge_coef.sort_values(by='Coef',ascending=False).head(10)
```

```python
r_top10=ridge_coef.sort_values(by='Coef',ascending=False).head(10)
plt.figure(figsize=(20,20))
plt.subplot(4,3,1)
sns.barplot(y = 'Feature', x='Coef', palette='Set1', data = r_top10)
plt.show()
```
